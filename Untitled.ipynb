```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import numpy as np


def plot_classification_results(g, x_min, x_max, y_min, y_max, title, h = .01):

    # Create color maps
    cmap_light = ListedColormap(['#FFAAAA', '#FFFFAA', '#AAAAFF'])
    
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))

    Z = g(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.figure(dpi=300)
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title(title)
    

def g(x):
    x1, x2 = x[:, 0], x[:, 1]
    z = (x1 - 0.5) + (x2 - 0.25) * 1j
    return np.floor((np.angle(z, deg=True) + 180) / 120)

plot_classification_results(g, -1, 2, -0.5, 1.5, "Decision Boundary")
```
